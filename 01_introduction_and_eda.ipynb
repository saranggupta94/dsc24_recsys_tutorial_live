### What is MovieLens?

MovieLens is a recommender system that was developed by GroupLens, a computer science research lab at the University of Minnesota. It recommends movies to its users based on their movie ratings. It is also a dataset that is widely used in research and teaching contexts.

### Chapter 1: Load Data and Exploratory Data Analysis (EDA)

```python
import warnings

# Suppress the specific FutureWarning
warnings.filterwarnings('ignore', category=FutureWarning)
pd.set_option('display.float_format', lambda x: '%.0f' % x)
```

```python
#Importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Reading ratings file
ratings = pd.read_csv('data-1m/ratings.csv', 
                    sep='\t', #Note that the separator here is "\t"
                    encoding='latin-1',
                    engine='python',
                    index_col=0
                     ) 

# Reading users file
users = pd.read_csv('data-1m/users.csv', 
                    sep='\t', #Note that the separator here is "\t"
                    encoding='latin-1',
                    engine='python',
                    index_col=0
                     )

# # Reading movies file
movies = pd.read_csv('data-1m/movies.csv', 
                    sep='\t', #Note that the separator here is "\t"
                    encoding='latin-1',
                    engine='python',
                    index_col=0
                     ) 
```

### Movies Dataset

```python
#Printing the head of the dataset
movies.head()
```

```python
#Inspecting if there are any nulls in the dataset
movies.info()
```

```python
# Import new libraries
%matplotlib inline
import wordcloud
from wordcloud import WordCloud, STOPWORDS

# Create a wordcloud of the movie titles
movies['title'] = movies['title'].fillna("").astype('str')
title_corpus = ' '.join(movies['title'])
title_wordcloud = WordCloud(stopwords=STOPWORDS, background_color='white', height=2000, width=4000).generate(title_corpus)

# Plot the wordcloud
plt.figure(figsize=(16,8))
plt.imshow(title_wordcloud)
plt.axis('off')
plt.show()
```

```python
import seaborn as sns
all_genres = movies['genres'].str.split('|').explode()

# Count the frequency of each genre
genre_counts = all_genres.value_counts()

# Create a figure with larger size
plt.figure(figsize=(12, 6))

# Create a bar plot using seaborn
sns.barplot(x=genre_counts.values, y=genre_counts.index, palette='viridis')

# Customize the plot
plt.title('Distribution of Movie Genres', fontsize=14, pad=15)
plt.xlabel('Number of Movies', fontsize=12)
plt.ylabel('Genre', fontsize=12)

# Add grid lines for better readability
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Set up the plotting style
# plt.style.use('seaborn-v0_8')  # or simply remove this line as we're using seaborn anyway
# sns.set_palette('viridis')

# Create a figure with subplots
fig = plt.figure(figsize=(20, 12))

# 1. Gender Distribution
plt.subplot(2, 2, 1)
gender_dist = users['gender'].value_counts()
sns.barplot(x=gender_dist.index, y=gender_dist.values)
plt.title('Gender Distribution')
plt.ylabel('Count')

# 3. Age Group Distribution
plt.subplot(2, 2, 2)
age_desc_dist = users['age_desc'].value_counts()
sns.barplot(x=age_desc_dist.values, y=age_desc_dist.index)
plt.title('Age Group Distribution')
plt.xlabel('Count')

# 4. Top 10 Occupations
plt.subplot(2, 2, 3)
occ_dist = users['occ_desc'].value_counts().head(10)
sns.barplot(x=occ_dist.values, y=occ_dist.index)
plt.title('Top 10 Occupations')
plt.xlabel('Count')

# 5. Age vs Occupation (Box Plot)
plt.subplot(2, 2, 4)
sns.boxplot(data=users, y='occ_desc', x='age', order=occ_dist.index)
plt.title('Age Distribution by Top Occupations')
plt.xlabel('Age')

plt.tight_layout()
plt.show()
```

## Age & Generation Distribution
1. **Digital Natives and Early Internet Users**
   - The largest group is 25-34 years (34.7% of users), followed by 35-44 years (19.8%)
   - Given that this is the MovieLens dataset, these age groups represent early adopters of online movie recommendation systems
   - These users were likely comfortable with technology and early internet services

2. **Student Representation**
   - Strong presence of students (18-24: 18.3%)
   - Combined with the occupation data showing students as the largest single group (759 users)
   - This suggests MovieLens had strong adoption in academic environments

## Professional Background Insights

3. **Tech-Savvy User Base**
   - High representation of technical professions:
     * Technicians/Engineers: 502 users
     * Programmers: 388 users
     * Combined with executives (679) and educators (528)
   - This suggests early MovieLens users were more technically inclined, which could influence movie preferences toward sci-fi and technical themes

4. **Creative Professionals**
   - Significant presence of creative professionals:
     * Writers: 281 users
     * Artists: 267 users
   - This demographic might have different movie preferences, possibly more oriented toward independent films and artistic content

5. **Academic Connection**
   - Strong academic presence:
     * Students: 759 users
     * Educators: 528 users
   - Total of 1,287 users (21.3%) connected to academia
   - This suggests MovieLens might have started with strong university connections

## Gender Distribution and Implications

6. **Notable Gender Imbalance**
   - Male: 71.7% (4,331 users)
   - Female: 28.3% (1,709 users)
   - This significant gender skew is typical of early internet services and tech platforms from this era
   - May have influenced the types of movies rated and recommended on the platform

## Demographic Implications for Movie Recommendations

7. **Target Audience Profile**
   - Primary: Tech-savvy males in their late 20s to early 40s
   - Secondary: Students and academics
   - This demographic mix suggests users who might be interested in:
     * Science fiction and technology-themed movies
     * Independent and art house films
     * Academic or intellectual content
     * Classic films and film history

8. **Professional Diversity**
   - The mix of technical and creative professionals suggests:
     * Diverse taste in movies
     * Potential for both mainstream and niche film interests
     * Higher likelihood of analytical approach to movie ratings

9. **Age-Based Content Preferences**
   - Working-age adults dominate (25-49 age groups)
   - Suggests users with:
     * Disposable income for entertainment
     * Interest in mature themes and complex narratives
     * Nostalgia for films from their youth (70s-90s)
